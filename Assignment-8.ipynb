```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
sns.set_style('whitegrid')
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
data
```

```python
data.describe()
```

```python
data['Sex'].unique()
```

```python
df = data.copy();
df
```

```python
# using group by method to calculate gender distribution

print("Number of Records grouped by Gender:\n ")
print(df.groupby(["Sex"])["PassengerId"].count())
```

```python
# count of records grouped by gender

sns.countplot(x = 'Sex', data = df)
```

```python
# using group by method to calculate survival distribution

print("Number of Records grouped by Survival record:\n ")
print(df.groupby(["Survived"])["PassengerId"].count())
```

```python
# pie chart distribution for count of survival

labels = ['Not Survived', 'Survived']
size = df['Survived'].value_counts()
explode = [0, 0.1]

plt.rcParams['figure.figsize'] = (6, 6)
plt.pie(size, explode = explode, labels = labels, shadow = False, autopct = '%.2f%%')
plt.title('Survival', fontsize = 20)
plt.legend()
plt.show()
```

```python
sns.catplot(x = "Sex", hue = "Survived", kind = "count", data = df)
```

A very low percentage of male passengers survived while higher percentage of female passengers survived.

```python
g = sns.FacetGrid(data = df, col = 'Sex')
g.map(sns.distplot, 'Age', kde = False)
```

Average age of male passengers ranges between 20 to 50.
Average age of femal passengers ranges between 20 to 40.

```python
sns.boxplot(x = 'Pclass', y = 'Age', data = df)
```

```python
sns.swarmplot(x = 'Pclass', y = 'Age', data = df)
```

```python
plt.figure(figsize = (10, 8))
sns.heatmap(data = df.corr(), annot = True, cmap = 'RdGy')
```

According to heatmap, Fare highly depends on the Class of the passengers.

```python
sns.jointplot(x = 'Fare', y = 'Age', data = df)
```

Highest prices of tickets can be seen for passengers aged between **20 to 50**.
