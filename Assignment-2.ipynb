```python
# columns having the data type as "float64"
cols = df_test.select_dtypes(['float64']).columns
cols
```

```python
# IQR value for math, writing and reading columns
from scipy import stats
dft = df_test.copy()
result = stats.iqr(dft[cols], axis = 0)
result
```

```python
dft.shape
```

```python
dft.drop(dft[dft['math score'] < (dft['math score'].quantile(0.25) - 1.5 * result[0])].index, inplace = True)
dft.drop(dft[dft['reading score'] < (dft['reading score'].quantile(0.25) - 1.5 * result[1])].index, inplace = True)
dft.drop(dft[dft['writing score'] < (dft['writing score'].quantile(0.25) - 1.5 * result[2])].index, inplace = True)
```

# Scaling data from range 0 to 10

```python
df_scaled = dft.copy()
features = df_scaled[cols]
```

```python
# Using MinMaxScaler to scale in between 0 to 10.
from sklearn.preprocessing import MinMaxScaler
# scale = (x - min) / (max - min)

scaler = MinMaxScaler(feature_range = (0, 10))
df_scaled[cols] = scaler.fit_transform(features.values)
```
